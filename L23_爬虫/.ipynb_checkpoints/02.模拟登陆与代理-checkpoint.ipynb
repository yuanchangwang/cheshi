```python
import requests
from urllib import request
from lxml import etree

requests = requests.Session()


home_url = "https://so.gushiwen.org/user/login.aspx?from=http://so.gushiwen.org/user/collect.aspx"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.75 Safari/537.36"
}

# 1.向登录页发送请求，获取页面源码数据，用来解析验证码图片，和登录必要的两个参数__VIEWSTATE和__VIEWSTATEGENERATOR
page_text = requests.get(url=home_url, headers=headers).text

# 2.解析验证码图片并下载到本地
tree = etree.HTML(page_text)
img_url = "https://so.gushiwen.org" + tree.xpath('//img[@id="imgCode"]/@src')[0]

# 这里会有问题，使用urlretreve方法下载的验证码图片无法实现登录成功, 所以要用原始的requests模块来实现验证码图片的下载
# request.urlretrieve(img_url, "img_code.jpg")
page_content = requests.get(url=img_url, headers=headers).content
with open("./img_code.jpg", 'wb') as f:
    f.write(page_content)

# 3.解析登录必要的两个参数
view_state = tree.xpath('//*[@id="__VIEWSTATE"]/@value')[0]
view_state_generator = tree.xpath('//*[@id="__VIEWSTATEGENERATOR"]/@value')[0]

# 4.验证码解析
img_code = get_code_img("./img_code.jpg", 1004)

# 5.封装模拟登录必要的参数
login_url = "https://so.gushiwen.org/user/login.aspx?from=http%3a%2f%2fso.gushiwen.org%2fuser%2fcollect.aspx"

form_data = {
    "__VIEWSTATE": view_state,
    "__VIEWSTATEGENERATOR": view_state_generator,
    "from": "http://so.gushiwen.org/user/collect.aspx",
    "email": "lxh661314@163.com",
    "pwd": "nihao123!",
    "code": img_code,
    "denglu": "登录"
}

# 6.实现模拟登录
page_text = requests.post(url=login_url, headers=headers, data=form_data).text

with open('./gushiwen.html', 'w', encoding="utf-8") as f:
    f.write(page_text)

# 总结:
# 1. urllib模块里面的request.urlretreve是不能够携带UA伪装的,需要使用requests模块来替代
# 2. 可以使用requests模块里面的Session对象来实现请求的发送，好处是可以自动携带并处理cookie
```

### 代理: 代理服务器
    - 快代理
    - 西祠代理
    - goubanjia
- 匿名度:
    - 透明: 对方服务器知道你使用了代理IP也知道你的真实IP
    - 匿名: 知道你使用了代理IP但是不知道你的真实IP
    - 高匿: 什么都不知道
- 类型:
    - http: 只可以发起http请求
    - https: 只可以发起https请求

```python
import requests
import random
from lxml import etree

url = "https://www.baidu.com/s?wd=ip"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.75 Safari/537.36"
}

proxy_ip_list = [
    {"https": "218.2.226.42:80"},
    {"https": "111.26.161.198:808"},
    {"https": "123.139.56.238:9999"}
]

page_text = requests.get(url=url, headers=headers, proxies=random.choice(proxy_ip_list)).text
tree = etree.HTML(page_text)
ip_data = tree.xpath('//span[@class="c-gap-right"]/text()')

print(ip_data)

# 总结:
# 1. 在requests发送请求时携带proxies参数就可以实现代理IP的应用
```

### cookie的应用和处理
    - cookie: 服务器端记录客户端的相关状态
    - 处理cookie的方式:
        - 手动处理: 不建议
        - 自动处理: 会话对象Session, 该对象可以像requests模块一样进行网络请求的发送(get, post)
          Session进行的请求发送可以自动携带和处理cookie

```python
# 后续所有的需求都使用session进行请求发送.
import json
import requests

session = requests.Session()

home_url = "https://xueqiu.com/"
url = "https://xueqiu.com/v4/statuses/public_timeline_by_category.json?since_id=-1&max_id=20352655&count=15&category=-1"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.75 Safari/537.36"
}

# 向首页发送get请求，获取cookie并保存在session里面
session.get(url=home_url, headers=headers)

# 向动态加载数据的ajax数据包的URL发送请求，并携带cookie
page_json = session.get(url=url, headers=headers).json()

for news_data in page_json["list"]:
#     detail_url = "https://xueqiu.com" + news_data["data"]["target"]
    detail_url = "https://xueqiu.com" + json.loads(news_data["data"])["target"]
    print(detail_url)

# 总结：
# 1. 向URL发送请求必须要携带cookie，因为有些网站会有cookie相关的检测机制
# 2. 有些json数据并不是真正的字典或列表，需要使用json模块进行序列化才可以得到真正的字典和列表
```

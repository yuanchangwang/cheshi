### selenium
- 概念: 用来完成浏览器自动化相关的操作，可以通过代码的形式制定一些基于浏览器自动化的相关操作(行为动作), 当代码执行后，浏览器就会自动触发相关的事件
- 环境安装:
    - pip install selenium
    - 下载对应浏览器的驱动程序
- 编码流程:
    - 导包: from selenium import webdriver
    - 实例化某一款浏览器对象
    - 制定相关的行为动作
    

```python
import time
from selenium import webdriver

bro = webdriver.Chrome(executable_path='./chromedriver.exe')
bro.get('https://www.baidu.com')
time.sleep(2)

# 定位到input输入框
input_ele = bro.find_element_by_id("kw")

# 输入刘德华关键字
input_ele.send_keys("刘德华")
time.sleep(2)

# 点击百度一下进行搜索
check_ele = bro.find_element_by_id("su")
check_ele.click()
time.sleep(2)

bro.close()
```

```python
# 使用selenium爬取雪球网更多数据
import time
from selenium import webdriver

bro = webdriver.Chrome(executable_path='./chromedriver.exe')
bro.get('https://xueqiu.com/')
time.sleep(2)

# 使用selenium实现滚轮往下拖动一个屏幕的高度
js = "window.scrollTo(0, document.body.scrollHeight)"
bro.execute_script(js)
time.sleep(2)
bro.execute_script(js)
time.sleep(2)
bro.execute_script(js)
time.sleep(2)
bro.execute_script(js)
time.sleep(2)

print(bro.page_source)

with open('./xueqiu.html', 'w', encoding='utf-8') as f:
    f.write(bro.page_source)

bro.close()
```

```python
# 谷歌无头浏览器
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# 创建一个参数对象，用来控制chrome以无界面模式打开
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')

bro = webdriver.Chrome(executable_path='./chromedriver.exe', chrome_options=chrome_options)
bro.get('https://www.baidu.com')
bro.save_screenshot("./1.jpg")
time.sleep(2)

# 定位到input输入框
input_ele = bro.find_element_by_id("kw")

# 输入刘德华关键字
input_ele.send_keys("刘德华")
time.sleep(2)
bro.save_screenshot("./2.jpg")

# 点击百度一下进行搜索
check_ele = bro.find_element_by_id("su")
check_ele.click()
time.sleep(2)

bro.close()
```

```python
# 动作链
import time
from selenium import webdriver
from selenium.webdriver import ActionChains
 
bro=webdriver.Chrome(executable_path='./chromedriver.exe')
bro.get("https://www.runoob.com/try/try.php?filename=jqueryui-api-droppable")

# 1.切换到iframe页面内部
bro.switch_to.frame('iframeResult')

# 2.定位源标签
source_tag = bro.find_element_by_id('draggable')

# 3.实例化一个动作链对象
action = ActionChains(bro)

# 4.制定动作链相关的操作
action.click_and_hold(source_tag)

for i in range(0,5):
    action.move_by_offset(20, 0)
    time.sleep(1)

# 5.执行动作链
action.perform()
time.sleep(2)

bro.close()
```

```python
# 动作链(完整的拖拽操作)
import time
from selenium import webdriver
from selenium.webdriver import ActionChains  # 动作链
from selenium.webdriver import ChromeOptions

bro = webdriver.Chrome(executable_path='./chromedriver.exe')
url = "http://www.runoob.com/try/try.php?filename=jqueryui-api-droppable"
bro.get(url=url)

# 如果定位的标签存在于iframe标签之中, 则必须经过switch_to操作再进行标签定位
bro.switch_to.frame("iframeResult")
source_tag = bro.find_element_by_id('draggable')
target_tag = bro.find_element_by_id('droppable')

# 实例化一个动作链对象
action = ActionChains(bro)
action.click_and_hold(source_tag)
time.sleep(1)

for i in range(0, 5):
    action.move_by_offset(20, 0)
action.drag_and_drop(source_tag, target_tag)

# 表示开始执行动作链
action.perform()
time.sleep(2)

bro.quit()
```
